### Local model via Ollama

```python
from shared.utils import chat

# Uses LLM_PROVIDER from .env (default: local / phi3:mini)
reply = chat('Explain what a Markov chain is in 2 sentences.')
print(reply)
```

### Switching to cloud mid-notebook

```python
from shared.utils import get_llm
from langchain_core.messages import HumanMessage

# Force cloud for this call only
llm_cloud = get_llm(provider='cloud')
response = llm_cloud.invoke([HumanMessage(content='What is entropy in thermodynamics?')])
print(response.content)
```

### Streaming output (good for long responses)

```python
llm = get_llm()  # local
for chunk in llm.stream('Count from 1 to 5 slowly.'):
    print(chunk.content, end='', flush=True)
```
